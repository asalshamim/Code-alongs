➊ Count numbers

```python
n= -10
while n <= 10:
    print(n, end =", ")
    n+= 1
```

⓶ Arithmetic sum

a)

```python
n=1
summa=0
while n<=100:
    if n==100:
        print(f"100=",end="")
    else:
        print(f"{n}+",end="")
    summa += n
    n+=1
    print(summa)
```

b)

```python
n=1
summa=0
while n<=99:
    if n==99:
        print(f"99=",end="")
    else:
        print(f"{n}+",end="")
    summa += n
    n+=1
    print(summa)
```
